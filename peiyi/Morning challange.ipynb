```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score, roc_curve, auc, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# Load the dataset as df
train_data_path = 'D:/University documents/Year 4/Data Science Challange/transactions_training_sept_oct_2023.csv'
df_train = pd.read_csv(train_data_path, sep=';', decimal=',')
df_train
```

```python
df_train = df_train.drop(columns=['ID_TRX', 'ID_CARD', 'DATETIME_GMT'])
df_train
```

```python
# For creating a validation dataset by randomly selecting 20% of the training data
val_df = df_train.sample(frac=0.2, random_state=42)
# The remaining 80% will be used for training
train_df = df_train.drop(val_df.index)
```

```python
train_df.isna().values.any()
```

```python
# The dataset is very imbalanced, with a majority of transactions being non-fraudulent (0) and a minority being fraudulent (1).
# We may count how many fraudulent transactions there are in the training set.
fraud_count = df_train['FLAG_FRAUD'].value_counts()
print(fraud_count)
```

```python
# A ratio between the number of non-fraudulent and fraudulent transactions, to the nearest integer.
ratio = fraud_count[0] // fraud_count[1]
ratio
```

#### Method 1: logistic regression for classification

Considering there are NA values, we may apply two methods:

1. Replace them with 0 (which might be heavily biased).

2. Drop the columns with a lot of NA values.

```python
# Method 1: replace NA values with 0
train_df = train_df.fillna(0)
val_df = val_df.fillna(0)
```

```python
# Define the feature data (X), feature names, and target data (y), for both training and valing data
X_train = train_df.iloc[:, :26].values  # values for training feature data
y_train = train_df.iloc[:, -1].values  # values for target data (training labels)
X_val = val_df.iloc[:, :26].values  # values for validation feature data
y_val = val_df.iloc[:, -1].values  # values for target data (validation

feature_names = list(df_train.columns[:26])  # 25 design properties

# Quick chek of the shape
X_train.shape, y_train.shape, X_val.shape, y_val.shape, feature_names
```

```python
# Fit a standard logistic regression model
lr_model = LogisticRegression(solver='lbfgs', max_iter=1000, random_state=42)
lr_model.fit(X_train, y_train)

# Make predictions with the standard model
y_pred_train_lr = lr_model.predict(X_train)
y_pred_val_lr = lr_model.predict(X_val)

# Evaluate the standard model
print(classification_report(y_train, y_pred_train_lr))
print(classification_report(y_val, y_pred_val_lr))
```

```python
# Fit a weighted logistic regression model
weights = {0: 1, 1: ratio}  # our weights would be the ratio of non-fraudulent to fraudulent transactions
weighted_lr_model = LogisticRegression(solver='lbfgs', max_iter=1000, random_state=42, class_weight=weights)
weighted_lr_model.fit(X_train, y_train)

# Make predictions with the weighted model
y_pred_train_weighted_lr = weighted_lr_model.predict(X_train)
y_pred_val_weighted_lr = weighted_lr_model.predict(X_val)

# Evaluate the weighted model
print(classification_report(y_train, y_pred_train_weighted_lr))
print(classification_report(y_val, y_pred_val_weighted_lr))
```
