```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("../data/transactions_training_sept_oct_2023.csv", sep=";", decimal=",")
```

```python
# Filter only rows with known labels
df = df_train.copy()

# Define your feature lists
numerical_cols = [col for col in df.columns if (
    col.startswith("Anomaly_amount") or
    col.startswith("Population_Anomaly") or
    col == "AMOUNT"
)]

categorical_cols = [col for col in df.columns if "FLAG_BEHAVIOUR_Anomaly" in col]

# --- 1. NUMERICAL COMPARISONS ---
for col in numerical_cols:
    plt.figure()
    sns.kdeplot(np.log(1+df[df['FLAG_FRAUD'] == 0][col]), label='Not Fraud (0)', fill=True, alpha=0.5)
    sns.kdeplot(np.log(1+df[df['FLAG_FRAUD'] == 1][col]), label='Fraud (1)', fill=True, alpha=0.5)
    plt.title(f"KDE Plot of {col} by Fraud Flag")
    plt.xlabel(col)
    plt.legend()
    plt.tight_layout()
    plt.show()

# --- 2. CATEGORICAL COMPARISONS ---
for col in categorical_cols:
    plt.figure()
    ctab = pd.crosstab(df[col], df['FLAG_FRAUD'], normalize='columns')  # normalize per class
    ctab.plot(kind='bar')
    plt.title(f"{col} Frequency by Fraud Flag")
    plt.ylabel("Proportion")
    plt.xlabel(col)
    plt.xticks(rotation=0)
    plt.legend(['Not Fraud (0)', 'Fraud (1)'])
    plt.tight_layout()
    plt.show()
```
